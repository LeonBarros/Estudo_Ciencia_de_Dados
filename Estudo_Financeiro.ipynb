```python
import pandas as pd
import matplotlib.pyplot as plt

# Lista com os arquivos mensais (CSV)
arquivos = ['NU_961313337_01MAI2025_31MAI2025.csv', 'NU_961313337_01JUN2025_30JUN2025.csv', 'NU_961313337_01JUL2025_31JUL2025.csv']

# Lê e junta todos os arquivos CSV
df_list = [pd.read_csv(arq) for arq in arquivos]
df = pd.concat(df_list, ignore_index=True)

# Converte a coluna 'Data' para datetime
df['Data'] = pd.to_datetime(df['Data'], dayfirst=True)

# Define a coluna 'Data' como índice
df.set_index('Data', inplace=True)

# Ordena por data
df.sort_index(inplace=True)

plt.figure(figsize=(14, 5))

# Exibe a série temporal
df['Valor'].plot(title="Série Temporal Financeira")
plt.xlabel("Data")
plt.ylabel("Valor")
plt.grid(True)
plt.show()
```

```python
# Somar gastos por dia (considerando 'Data' como coluna)
df_dia = df.groupby('Data')['Valor'].sum().reset_index()

# Filtra só os dias com gasto > 0
df_gastos = df_dia[df_dia['Valor'] < 0]

print(df_gastos)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 5))
plt.plot(df_gastos['Data'], df_gastos['Valor'], marker='o', linestyle='-')
plt.title("Gastos Diários (Só dias com gasto > 0)")
plt.xlabel("Data")
plt.ylabel("Valor")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# --- Dados ordenados ---
categorias = ['Fatura', 'Alimentação', 'Lazer', 'Gastos Fixos', 'Mercado', 'Transporte']
valores = [54.44, 26.32, 10.80, 2.99, 2.80, 2.65]  # já em %

# Paleta suave (padronizada)
cores_padrao = {
    'Fatura': '#ff9999',
    'Alimentação': '#66b3ff',
    'Lazer': '#99ff99',
    'Gastos Fixos': '#ffcc99',
    'Mercado': '#c2c2f0',
    'Transporte': '#ffb3e6'
}

# Lista de cores na mesma ordem das categorias
cores = [cores_padrao[c] for c in categorias]

# Criar gráfico de barras
plt.figure(figsize=(8, 6))
plt.bar(categorias, valores, color=cores, edgecolor='white', linewidth=1)

# Título e rótulos
plt.title('Distribuição Financeira - Maio (%)', fontsize=14, fontweight='bold')
plt.ylabel('Percentual (%)')
plt.xticks(rotation=30, ha='right')

# Exibir valores acima das barras com símbolo %
for i, valor in enumerate(valores):
    plt.text(i, valor + 0.5, f'{valor:.2f}%', ha='center', fontsize=9)

plt.tight_layout()
plt.show()
```
