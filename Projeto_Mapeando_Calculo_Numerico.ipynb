## Parte I: Números Binarios

### Conversão de Números: Binaria e Decimal

```python
# Conversão de Decimais para o Binário (Exibindo Cálculos)

def decimal_para_binario_didatico(n):
    print(f"\nConvertendo {n} para binário:")
    potencias = []
    expoente = 0
    while 2 ** expoente <= n:
        potencias.append(expoente)
        expoente += 1
    potencias = potencias[::-1]  # do maior para o menor

    binario = ''
    expressao = []
    for e in potencias:
        p = 2 ** e
        if n >= p:
            binario += '1'
            expressao.append(f"1×2^{e}")
            print(f"{n} ≥ 2^{e} = {p} → 1 (resta {n - p})")
            n -= p
        else:
            binario += '0'
            expressao.append(f"0×2^{e}")
            print(f"{n} < 2^{e} = {p} → 0")

    print("\nExpressão final:")
    print(" + ".join(expressao))
    print(f"Resultado final em binário: {binario}")
    return binario

# Exemplo
numero = int(input("Digite um número inteiro: "))
decimal_para_binario_didatico(numero)
```

```python
# Conversão de Números Binários para Decimais (Exibindo Calcúlos)

def binario_para_decimal_inteiro(binario_str):
    binario_str = binario_str.strip()
    binario_str = binario_str[::-1]  # inverte para facilitar: da direita para a esquerda
    total = 0
    print(f"\nConvertendo {binario_str[::-1]} (binário) para decimal:")

    for i, bit in enumerate(binario_str):
        valor = int(bit) * (2 ** i)
        print(f"{bit} × 2^{i} = {valor}")
        total += valor

    print(f"Resultado final: {total}")
    return total

# Exemplo
binario = input("Digite a parte inteira em binário (ex: 1101): ")
binario_para_decimal_inteiro(binario)
```

```python
# Aritmética em Ponto Flutuante

def decimal_para_binario_inteira(n):
    print(f"\nConvertendo {n} (parte inteira) para binário:")
    if n == 0:
        print("0 × 2^0 = 0")
        return "0"

    potencias = []
    expoente = 0
    while 2 ** expoente <= n:
        potencias.append(expoente)
        expoente += 1
    potencias = potencias[::-1]

    binario = ''
    expressao = []
    for e in potencias:
        p = 2 ** e
        if n >= p:
            binario += '1'
            expressao.append(f"1×2^{e}")
            print(f"{n} ≥ 2^{e} = {p} → 1 (resta {n - p})")
            n -= p
        else:
            binario += '0'
            expressao.append(f"0×2^{e}")
            print(f"{n} < 2^{e} = {p} → 0")

    print("Expressão:", " + ".join(expressao))
    print(f"Parte inteira em binário: {binario}")
    return binario

def decimal_para_binario_fracionaria(frac, max_bits=10):
    print(f"\nConvertendo {frac} (parte fracionária) para binário:")
    binario = ''
    expressao = []
    for i in range(1, max_bits + 1):
        frac *= 2
        bit = int(frac)
        binario += str(bit)
        expressao.append(f"{bit}×2^-{i}")
        print(f"{frac} → {bit} × 2^(-{i})")
        frac -= bit
        if frac == 0:
            break

    print("Expressão:", " + ".join(expressao))
    print(f"Parte fracionária em binário: {binario}")
    return binario

# Entrada do usuário
parte_inteira = int(input("Digite a parte inteira (ex: 13): "))
entrada_fracionada = input("Digite a parte fracionária (ex: 0.625): ")

# Validação de entrada com vírgula
if ',' in entrada_fracionada:
    print("\nPor favor, use ponto em vez de vírgula para decimais.")
    print("Exemplo correto: 0.625  Errado: 0,625")
else:
    try:
        parte_fracionada = float(entrada_fracionada)

        # Conversão
        bin_inteira = decimal_para_binario_inteira(parte_inteira)
        bin_fracionada = decimal_para_binario_fracionaria(parte_fracionada)

        # Resultado final
        print("\nResultado final:")
        print(f"Binário completo: {bin_inteira}.{bin_fracionada}")

    except ValueError:
        print("\nEntrada inválida. Por favor, insira um número decimal válido.")
```

```python
# Aritmética em ponto flutuante

def simular_ponto_flutuante(numero):
    print(f"\nSimulando ponto flutuante para o número: {numero}")

    expoente = 0
    mantissa = abs(numero)

    # normalizar: trazer a mantissa para [1, 2)
    while mantissa >= 2:
        mantissa /= 2
        expoente += 1
    while mantissa < 1 and mantissa != 0:
        mantissa *= 2
        expoente -= 1

    sinal = "-" if numero < 0 else "+"

    print(f"Forma normalizada: {sinal}{mantissa:.5f} × 2^{expoente}")


# Entrada do usuário
entrada = input("Digite um número decimal (use ponto: ex: 6.25): ")

if ',' in entrada:
    print("Por favor, use ponto no lugar da vírgula. Exemplo correto: 3.5 (errado: 3,5)")
else:
    try:
        numero = float(entrada)
        simular_ponto_flutuante(numero)
    except ValueError:
        print("Entrada inválida. Digite um número real válido.")
```

```python
 # Aritmética em Ponto Flutuante (Exibição Computacional)

def float_para_binario_simples(numero):
    sinal = '0' if numero >= 0 else '1'
    n = abs(numero)
    e = 0

    while n >= 2:
        n /= 2
        e += 1
    while n < 1 and n != 0:
        n *= 2
        e -= 1

    expoente_polarizado = e + 127
    mantissa = n - 1

    print("\n➤ 1º Passo: Normalização do número:")
    print(f"Número normalizado: 1.{str(mantissa)[2:]} × 2^{e}")

    print("\n➤ 2º Passo: Cálculo do expoente Polarizado em precisão Simples:")
    print(f"{e} + 127 = {expoente_polarizado}")
    print(f"Expoente (binário): {expoente_polarizado:08b}")

    mantissa_bits = ''
    temp = mantissa
    for _ in range(23):
        temp *= 2
        bit = int(temp)
        mantissa_bits += str(bit)
        temp -= bit

    print(f"Bits da mantissa: {mantissa_bits}")

    return sinal, f"{expoente_polarizado:08b}", mantissa_bits

def float_para_binario_dupla(numero):
    sinal = '0' if numero >= 0 else '1'
    n = abs(numero)
    e = 0

    while n >= 2:
        n /= 2
        e += 1
    while n < 1 and n != 0:
        n *= 2
        e -= 1

    expoente_polarizado = e + 1023
    mantissa = n - 1

    print("\n➤ 1º Passo: Normalização do número:")
    print(f"Número normalizado: 1.{str(mantissa)[2:]} × 2^{e}")

    print("\n➤ 2º Passo: Cálculo do expoente Polarizado em precisão Dupla:")
    print(f"{e} + 1023 = {expoente_polarizado}")
    print(f"Expoente (binário): {expoente_polarizado:011b}")

    mantissa_bits = ''
    temp = mantissa
    for _ in range(52):
        temp *= 2
        bit = int(temp)
        mantissa_bits += str(bit)
        temp -= bit

    print(f"Bits da mantissa: {mantissa_bits}")

    return sinal, f"{expoente_polarizado:011b}", mantissa_bits

print("Cálculo do expoente Polarizado em precisão Dupla ou Simples:")
numero = float(input("Digite um número decimal: "))
precisao = input("Escolha a precisão: 0 para simples, 1 para dupla: ").strip()

if precisao == '0':
    sinal, expoente_bits, mantissa_bits = float_para_binario_simples(numero)
    print(f"\nResultado final (float32):")
    print(f"Sinal: {sinal}")
    print(f"Expoente: {expoente_bits}")
    print(f"Mantissa: {mantissa_bits}")
    print(f"Juntando tudo: {sinal + expoente_bits + mantissa_bits}")

elif precisao == '1':
    sinal, expoente_bits, mantissa_bits = float_para_binario_dupla(numero)
    print(f"\nResultado final (float64):")
    print(f"Sinal: {sinal}")
    print(f"Expoente: {expoente_bits}")
    print(f"Mantissa: {mantissa_bits}")
    print(f"Juntando tudo: {sinal + expoente_bits + mantissa_bits}")

else:
    print("Opção inválida")
```

```python
# Zero de funções Verificando a Existencia de Raiz

def f(x, funcao):
    return eval(funcao, {"x": x})

funcao = input("Digite a função em x (ex: x**3 - x - 2): ")
a = float(input("Digite o início do intervalo (a): "))
b = float(input("Digite o fim do intervalo (b): "))

fa = f(a, funcao)
fb = f(b, funcao)

print(f"f({a}) = {fa}")
print(f"f({b}) = {fb}")

if fa * fb < 0:
    print("Existe uma raiz no intervalo.")
elif fa == 0 or fb == 0:
    print("Um dos extremos é raiz exata.")
else:
    print("Não há garantia de raiz no intervalo.")
```

```python
# Zero de funções Metódo de Confinamento por Bisseção

def f(x, funcao):
    return eval(funcao, {"x": x})

funcao = input("Digite a função em x (ex: x**3 - x - 2): ")

a = float(input("Digite o início do intervalo (a): "))
b = float(input("Digite o fim do intervalo (b): "))
tol = float(input("Digite a tolerância (ex: 0.0001): "))
max_iter = int(input("Número máximo de iterações: "))

fa = f(a, funcao)
fb = f(b, funcao)

if fa * fb > 0:
    print("Não há mudança de sinal no intervalo. Não é possível aplicar a bisseção.")
else:
    for i in range(1, max_iter + 1):
        m = (a + b) / 2
        fm = f(m, funcao)

        print(f"Iteração {i}: m = {m}, f(m) = {fm}")

        if abs(fm) < tol or (b - a) / 2 < tol:
            print(f"\nRaiz aproximada encontrada: {m}")
            break

        if fa * fm < 0:
            b = m
            fb = fm
        else:
            a = m
            fa = fm
    else:
        print("\nNão convergiu dentro do número máximo de iterações.")
```

```python
# Zero de Funções pelo metódo de confinamento por cordas

def f(x, funcao):
    return eval(funcao, {"x": x})

funcao = input("Digite a função em x (ex: x**3 - x - 2): ")
a = float(input("Digite o valor de a (início do intervalo): "))
b = float(input("Digite o valor de b (fim do intervalo): "))
n = int(input("Digite o número de iterações: "))

fa = f(a, funcao)
fb = f(b, funcao)

if fa * fb > 0:
    print("A função não muda de sinal no intervalo. Não é possível aplicar o método.")
else:
    for i in range(1, n + 1):
        x = (a * fb - b * fa) / (fb - fa)
        fx = f(x, funcao)

        print(f"Iteração {i}: x = {x}, f(x) = {fx}")

        if fa * fx < 0:
            b = x
            fb = fx
        else:
            a = x
            fa = fx

    print(f"\nRaiz aproximada após {n} iterações: {x}")
```

```python
# Zero de funções usando Método Aberto por Newton

def f(x, funcao):
    return eval(funcao, {"x": x})

# Entrada
funcao = input("Digite a função f(x) (ex: x**3 - x - 2): ")
derivada = input("Digite a derivada f'(x) (ex: 3*x**2 - 1): ")
x = float(input("Digite o chute inicial (x0): "))
tol = float(input("Digite a tolerância (ex: 0.0001): "))
max_iter = int(input("Número máximo de iterações: "))

for i in range(1, max_iter + 1):
    fx = f(x, funcao)
    dfx = f(x, derivada)

    if dfx == 0:
        print("Derivada nula. Método falhou.")
        break

    x_novo = x - fx / dfx
    erro = abs(x_novo - x)

    print(f"Iteração {i}: x = {x_novo}, f(x) = {fx}, erro = {erro}")

    if erro < tol:
        print(f"\nRaiz aproximada encontrada: {x_novo}")
        break

    x = x_novo
else:
    print("\nNão convergiu dentro do número máximo de iterações.")
```

```python
# Zero de funções usando Método Aberto por Secante

def f(x, funcao):
    return eval(funcao, {"x": x})

# Entradas
funcao = input("Digite a função f(x) (ex: x**3 - x - 2): ")
x0 = float(input("Digite o primeiro chute inicial (x0): "))
x1 = float(input("Digite o segundo chute inicial (x1): "))
tol = float(input("Digite a tolerância (ex: 0.0001): "))
max_iter = int(input("Número máximo de iterações: "))

for i in range(1, max_iter + 1):
    fx0 = f(x0, funcao)
    fx1 = f(x1, funcao)

    if fx1 - fx0 == 0:
        print("Divisão por zero. Método falhou.")
        break

    x2 = x1 - fx1 * (x1 - x0) / (fx1 - fx0)
    erro = abs(x2 - x1)

    print(f"Iteração {i}: x = {x2}, f(x) = {f(x2, funcao)}, erro = {erro}")

    if erro < tol:
        print(f"\nRaiz aproximada encontrada: {x2}")
        break

    x0, x1 = x1, x2
else:
    print("\nNão convergiu dentro do número máximo de iterações.")
```

```python
# Gerador de Matrizes

import numpy as np

def gerar_matriz_aleatoria():
    linhas = int(input("Número de linhas: "))
    colunas = int(input("Número de colunas: "))

    # Escolher intervalo de valores
    print("\nVocê deseja que os valores da matriz sejam:")
    print("0 - Baixos (entre -5 e 5)")
    print("1 - Altos (entre -100 e 100)")
    intervalo = input("Sua escolha (0 ou 1): ").strip()

    if intervalo == "0":
        minimo, maximo = -5, 5
    elif intervalo == "1":
        minimo, maximo = -100, 100
    else:
        print("Escolha inválida. Usando padrão entre -10 e 10.")
        minimo, maximo = -10, 10

    # Escolher tipo dos valores
    print("\nVocê deseja que os valores sejam:")
    print("0 - Inteiros")
    print("1 - Decimais (ponto flutuante)")
    tipo_valor = input("Sua escolha (0 ou 1): ").strip()

    if tipo_valor == "0":
        matriz = np.random.randint(minimo, maximo + 1, size=(linhas, colunas))
    elif tipo_valor == "1":
        matriz = np.random.uniform(minimo, maximo, size=(linhas, colunas))
    else:
        print("Escolha inválida. Usando padrão decimal.")
        matriz = np.random.uniform(minimo, maximo, size=(linhas, colunas))

    print("\nMatriz gerada:")
    for i in range(linhas):
        linha = matriz[i]
        if tipo_valor == "0":
            linha_formatada = [str(int(num)) for num in linha]
        else:
            linha_formatada = ["{:.2f}".format(num) for num in linha]
        print("[" + ", ".join(linha_formatada) + "]")

    return matriz

# Executar o gerador
matriz = gerar_matriz_aleatoria()
```

```python
# Verificador de inversa de uma matriz

import numpy as np

# Entrada do usuário
n = int(input("Informe a ordem da matriz quadrada: "))
print("\nDigite os valores da matriz (linha por linha, separados por vírgula):")

matriz = []
for i in range(n):
    entrada = input(f"Linha {i+1}: ")
    valores = [float(x.replace(",", ".")) for x in entrada.split(",")]
    if len(valores) != n:
        print("Erro: número de valores diferente da ordem da matriz.")
        exit()
    matriz.append(valores)

# Converter para array numpy
A = np.array(matriz)

# Verificar se a matriz é invertível
det = np.linalg.det(A)

if np.isclose(det, 0):
    print("\nA matriz não possui inversa (determinante = 0).")
else:
    A_inv = np.linalg.inv(A)
    print("\nMatriz inversa:")
    for linha in A_inv:
        print("  ".join(f"{x:.5f}" for x in linha))
```

```python
# Trabalhando Sistemas de Equações Lineares

def eliminacao_gauss(A, b):
    n = len(b)

    for k in range(n - 1):
        for i in range(k + 1, n):
            if A[k][k] == 0:
                print("Divisão por zero. Método interrompido.")
                return None
            m = A[i][k] / A[k][k]
            for j in range(k, n):
                A[i][j] -= m * A[k][j]
            b[i] -= m * b[k]

    x = [0] * n
    for i in range(n - 1, -1, -1):
        soma = sum(A[i][j] * x[j] for j in range(i + 1, n))
        x[i] = (b[i] - soma) / A[i][i]

    return x


# Entrada do usuário
n = int(input("Quantas variáveis/equações o sistema possui? "))

# Aviso didático ao usuário
print("\n--- INSTRUÇÕES ---")
print("Você montará a matriz aumentada de um sistema linear.")
print("A ordem da entrada será:")
print("1️⃣  Os coeficientes kij de cada linha (separados por vírgula).")
print("2️⃣  Os nomes das variáveis (mostrados na estrutura).")
print("3️⃣  Os termos independentes bi, um por linha.\n")

# Mostrar estrutura simbólica da matriz aumentada
print("Estrutura simbólica da matriz aumentada:")
for i in range(n):
    linha = "  ".join([f"k{i}{j}·x{j+1}" for j in range(n)])
    print(f"[ {linha} ] | b{i}")

# Entrada dos coeficientes kij
print("\n Agora digite os coeficientes kij de cada linha, separados por vírgula:")
A = []
for i in range(n):
    entrada = input(f"Coeficientes da linha {i+1}: ").strip()
    valores = [float(v.replace(",", ".")) for v in entrada.split(",")]
    if len(valores) != n:
        print(f" A linha {i+1} deve conter exatamente {n} coeficientes.")
        exit()
    A.append(valores)

# Entrada dos termos independentes bi
print("\n Agora insira os termos independentes (b):")
b = []
for i in range(n):
    termo = input(f"b{i} (linha {i+1}): ").strip()
    b.append(float(termo.replace(",", ".")))

# Resolver o sistema
solucao = eliminacao_gauss(A, b)

# Mostrar resultado
if solucao:
    print("\n Solução do sistema:")
    for i, xi in enumerate(solucao):
        print(f"x{i+1} = {xi}")
```

```python
# Usando o Método de Gauss-Jacobi para Sistema de Equações Lineares

def gauss_jacobi(A, b, chute_inicial, tolerancia, max_iter):
    n = len(A)
    x_atual = chute_inicial[:]

    for iteracao in range(max_iter):
        x_novo = [0] * n
        for i in range(n):
            soma = sum(A[i][j] * x_atual[j] for j in range(n) if j != i)
            x_novo[i] = (b[i] - soma) / A[i][i]

        # Verifica a convergência
        erro = max(abs(x_novo[i] - x_atual[i]) for i in range(n))
        if erro < tolerancia:
            print(f"\nConvergiu em {iteracao+1} iterações.")
            return x_novo

        x_atual = x_novo[:]

    print("\nNão convergiu no número máximo de iterações.")
    return x_atual

# Entrada de dados
n = int(input("Quantas variáveis/equações o sistema possui? "))

print("\nDigite os coeficientes da matriz A (linha por linha, separados por vírgula):")
A = []
for i in range(n):
    linha = input(f"Linha {i+1}: ")
    A.append([float(x.replace(",", ".")) for x in linha.split(",")])

print("\nDigite os termos independentes (vetor b):")
b = []
for i in range(n):
    valor = input(f"b{i}: ")
    b.append(float(valor.replace(",", ".")))

print("\nDigite o chute inicial (valores separados por vírgula):")
chute_inicial = [float(x.replace(",", ".")) for x in input("x0: ").split(",")]

tolerancia = float(input("Digite a tolerância desejada: ").replace(",", "."))
max_iter = int(input("Digite o número máximo de iterações: "))

# Execução do método - guardando resultado em sol_jacobi
sol_jacobi = gauss_jacobi(A, b, chute_inicial, tolerancia, max_iter)

# Resultado
print("\nSolução aproximada pelo Gauss-Jacobi:")
for i, xi in enumerate(sol_jacobi):
    print(f"x{i+1} = {xi:.6f}")
```

```python
# Usando agora o Método de Gauss-Seidel

def gauss_seidel(A, b, chute_inicial, tolerancia, max_iter):
    n = len(A)
    x = chute_inicial[:]

    for iteracao in range(max_iter):
        x_novo = x[:]
        for i in range(n):
            soma1 = sum(A[i][j] * x_novo[j] for j in range(i))       # usa valor já atualizado
            soma2 = sum(A[i][j] * x[j] for j in range(i+1, n))       # usa valor antigo
            x_novo[i] = (b[i] - soma1 - soma2) / A[i][i]

        erro = max(abs(x_novo[i] - x[i]) for i in range(n))
        if erro < tolerancia:
            print(f"\nConvergiu em {iteracao+1} iterações.")
            return x_novo

        x = x_novo[:]

    print("\nNão convergiu no número máximo de iterações.")
    return x

# Entrada de dados
n = int(input("Quantas variáveis/equações o sistema possui? "))

print("\nDigite os coeficientes da matriz A (linha por linha, separados por vírgula):")
A = []
for i in range(n):
    linha = input(f"Linha {i+1}: ")
    A.append([float(x.replace(',', '.')) for x in linha.split(",")])

print("\nDigite os termos independentes (vetor b):")
b = []
for i in range(n):
    valor = input(f"b{i}: ")
    b.append(float(valor.replace(',', '.')))

print("\nDigite o chute inicial (valores separados por vírgula):")
chute_inicial = [float(x.replace(',', '.')) for x in input("x0: ").split(",")]

tolerancia = float(input("Digite a tolerância desejada: ").replace(",", "."))
max_iter = int(input("Digite o número máximo de iterações: "))

# Execução do método - guardando resultado em sol_seidel
sol_seidel = gauss_seidel(A, b, chute_inicial, tolerancia, max_iter)

# Resultado
print("\nSolução aproximada pelo Gauss-Seidel:")
for i, xi in enumerate(sol_seidel):
    print(f"x{i+1} = {xi:.6f}")
```

```python
# Condição de Convergência Pelo Critério de Linhas

def criterio_linhas(A):
    n = len(A)
    print("\nCritério das Linhas (αᵢ = soma dos outros / |aᵢᵢ|):\n")

    dominante = True

    for i in range(n):
        aii = abs(A[i][i])
        soma_outros = sum(abs(A[i][j]) for j in range(n) if j != i)

        print(f"α{i+1} = (", end="")
        partes = [f"|a{i+1}{j+1}|" for j in range(n) if j != i]
        print(" + ".join(partes) + f") / |a{i+1}{i+1}|")

        print(f"    = ({' + '.join(str(abs(A[i][j])) for j in range(n) if j != i)}) / {aii}")
        print(f"    = {soma_outros} / {aii} = {soma_outros / aii:.4f}")

        if soma_outros / aii < 1:
            print("→ Satisfaz: α < 1\n")
        else:
            dominante = False
            print("→ Não satisfaz: α ≥ 1\n")

    if dominante:
        print(" A matriz satisfaz o Critério das Linhas. Convergência garantida.\n")
    else:
        print(" A matriz NÃO satisfaz o Critério das Linhas. A convergência NÃO é garantida.\n")

# Entrada da matriz
n = int(input("Quantas variáveis/equações o sistema possui? "))

print("\nDigite os coeficientes da matriz A (uma linha por vez, separados por vírgula):")
A = []
for i in range(n):
    linha = input(f"Linha {i+1}: ")
    A.append([float(x.replace(",", ".")) for x in linha.split(",")])

# Verificação
criterio_linhas(A)
```

```python
# Condição de Convergência Pelo Critério de Colunas

def criterio_colunas(A):
    n = len(A)
    print("\nCritério das Colunas (αⱼ = soma dos outros / |aⱼⱼ|):\n")

    dominante = True

    for j in range(n):
        ajj = abs(A[j][j])
        soma_outros = sum(abs(A[i][j]) for i in range(n) if i != j)

        print(f"α{j+1} = (", end="")
        partes = [f"|a{i+1}{j+1}|" for i in range(n) if i != j]
        print(" + ".join(partes) + f") / |a{j+1}{j+1}|")

        print(f"    = ({' + '.join(str(abs(A[i][j])) for i in range(n) if i != j)}) / {ajj}")
        print(f"    = {soma_outros} / {ajj} = {soma_outros / ajj:.4f}")

        if soma_outros / ajj < 1:
            print("-> Satisfaz: α < 1\n")
        else:
            dominante = False
            print("-> Não satisfaz: α >= 1\n")

    if dominante:
        print("A matriz satisfaz o Critério das Colunas. Convergência garantida.\n")
    else:
        print("A matriz NÃO satisfaz o Critério das Colunas. Convergência NÃO garantida.\n")

# Entrada da matriz
n = int(input("Quantas variáveis/equações o sistema possui? "))

print("\nDigite os coeficientes da matriz A (uma linha por vez, separados por vírgula):")
A = []
for i in range(n):
    linha = input(f"Linha {i+1}: ")
    A.append([float(x.replace(",", ".")) for x in linha.split(",")])

# Verificação
criterio_colunas(A)
```

```python
# Condição de Convergência Pelo Critério de Sassenfeld

def criterio_sassenfeld(A):
    n = len(A)
    beta = [0] * n
    print("\nCálculo do Critério de Sassenfeld:\n")

    for i in range(n):
        soma = 0
        for j in range(n):
            if j < i:
                soma += abs(A[i][j]) * beta[j]  # usa β_j já calculado
            elif j > i:
                soma += abs(A[i][j]) * 1       # assume β_j = 1 para j > i

        beta[i] = soma / abs(A[i][i])

        # Mostrar passo a passo
        partes = []
        for j in range(n):
            if j != i:
                mult = f"β{j+1}" if j < i else "1"
                partes.append(f"|a{i+1}{j+1}|×{mult}")
        expressao = " + ".join(partes) if partes else "0"
        print(f"β{i+1} = ( {expressao} ) / |a{i+1}{i+1}| = {soma:.4f} / {abs(A[i][i])} = {beta[i]:.4f}")

    max_beta = max(beta)
    print(f"\nMaior βi = {max_beta:.4f}")

    if max_beta < 1:
        print("A matriz satisfaz o Critério de Sassenfeld. Convergência garantida.\n")
    else:
        print("A matriz NÃO satisfaz o Critério de Sassenfeld. Convergência NÃO garantida.\n")

# Entrada da matriz
n = int(input("Quantas variáveis/equações o sistema possui? "))

print("\nDigite os coeficientes da matriz A (uma linha por vez, separados por vírgula):")
A = []
for i in range(n):
    linha = input(f"Linha {i+1}: ")
    A.append([float(x.replace(',', '.')) for x in linha.split(',')])

criterio_sassenfeld(A)
```

```python
# Método de Newton Para a Solução de Sistemas de Equações Não Lineares

import sympy as sp
import numpy as np

# Definindo variáveis simbólicas
x, y = sp.symbols('x y')

# Entrada das funções via input
print("Insira as funções do sistema (use ** para potências e funções do numpy como exp, cos, log, etc.):")
f1_str = input("f1(x, y) = ")
f2_str = input("f2(x, y) = ")

# Transforma em expressões simbólicas
f1 = sp.sympify(f1_str)
f2 = sp.sympify(f2_str)

# Calcula a Jacobiana simbolicamente
J = sp.Matrix([[sp.diff(f1, x), sp.diff(f1, y)],
               [sp.diff(f2, x), sp.diff(f2, y)]])

# Converte para funções numéricas com backend numpy
f1_func = sp.lambdify((x, y), f1, "numpy")
f2_func = sp.lambdify((x, y), f2, "numpy")
J_func  = sp.lambdify((x, y), J, "numpy")

# Entrada dos dados iniciais
x0 = float(input("Chute inicial para x: "))
y0 = float(input("Chute inicial para y: "))
tol = float(input("Tolerância desejada (ex: 0.001): "))
max_iter = int(input("Número máximo de iterações: "))

# Método de Newton para sistemas
for k in range(1, max_iter + 1):
    F = np.array([f1_func(x0, y0), f2_func(x0, y0)], dtype=float)
    J_eval = np.array(J_func(x0, y0), dtype=float)

    try:
        delta = np.linalg.solve(J_eval, -F)
    except np.linalg.LinAlgError:
        print("Jacobiana singular ou mal-condicionada.")
        break

    dx, dy = delta
    x1, y1 = x0 + dx, y0 + dy

    # Erro relativo (considerando divisão segura)
    erro_x = abs((x1 - x0) / x1) if x1 != 0 else abs(x1 - x0)
    erro_y = abs((y1 - y0) / y1) if y1 != 0 else abs(y1 - y0)
    erro = max(erro_x, erro_y)

    print(f"\nIteração {k}:")
    print(f"  x = {x1:.6f}, y = {y1:.6f}")
    print(f"  Δx = {dx:.2e}, Δy = {dy:.2e}")
    print(f"  Erro relativo máximo: {erro:.2e}")

    if erro < tol:
        print(f"\nConvergiu em {k} iterações.")
        print(f"Solução: x = {x1:.6f}, y = {y1:.6f}")
        break

    x0, y0 = x1, y1
else:
    print("\nNão convergiu no número máximo de iterações.")
```

```python
# Resolvendo Sis. Eq. Não Lineares pelo Método de Newton-Raphson Modificado

import sympy as sp
import numpy as np

# Definir número de variáveis/equações
n = int(input("Número de variáveis/equações: "))
vars_symbols = sp.symbols(f'x1:{n+1}')  # x1, x2, ..., xn

# Entrada das funções
print(f"Insira as {n} funções do sistema usando as variáveis x1, x2, ..., xn:")
funcoes = []
for i in range(n):
    expr_str = input(f"f{i+1} = ")
    expr_str = expr_str.replace("np.", "")  # remove np. se houver
    funcoes.append(sp.sympify(expr_str))

# Jacobiana simbólica
J = sp.Matrix(n, n, lambda i, j: sp.diff(funcoes[i], vars_symbols[j]))

# Converter funções para funções numéricas
funcoes_num = [sp.lambdify(vars_symbols, f, "numpy") for f in funcoes]

# Entrada chute inicial
print(f"Insira o chute inicial para as variáveis x1 a x{n} separados por vírgula:")
x0_str = input()
x0_list = [float(v.replace(",", ".")) for v in x0_str.split(",")]
x0 = np.array(x0_list)

# Avaliar Jacobiana no chute inicial e fixar
J_eval = np.array(sp.lambdify(vars_symbols, J, "numpy")(*x0), dtype=float)

# Entrada da tolerância e máximo de iterações
tol = float(input("Tolerância desejada: ").replace(",", "."))
max_iter = int(input("Número máximo de iterações: "))

for k in range(1, max_iter + 1):
    F = np.array([f(*x0) for f in funcoes_num], dtype=float)

    try:
        delta = np.linalg.solve(J_eval, -F)
    except np.linalg.LinAlgError:
        print(f"Jacobiana singular ou mal-condicionada.")
        break

    x1 = x0 + delta

    erros = np.abs((x1 - x0) / np.where(x1 != 0, x1, 1))
    erro = np.max(erros)

    print(f"\nIteração {k}:")
    print("  Valores atuais:", ", ".join(f"x{i+1} = {v:.6f}" for i, v in enumerate(x1)))
    print("  Δ:", ", ".join(f"{d:.2e}" for d in delta))
    print(f"  Erro relativo máximo: {erro:.2e}")

    if erro < tol:
        print(f"\nConvergiu em {k} iterações.")
        print("Solução aproximada:")
        for i, val in enumerate(x1):
            print(f"x{i+1} = {val:.6f}")
        break

    x0 = x1
else:
    print("\nNão convergiu no número máximo de iterações.")
```

```python
# Resolvendo SE Não Lineares pelo Método da Secante

import sympy as sp
import numpy as np

# Variáveis simbólicas
x_sym, y_sym = sp.symbols('x y')

# Entrada das funções (apenas as expressões, sem "f1 =" etc)
print("Insira as duas funções do sistema com variáveis x e y:")
f1_str = input("f1 = ").replace("np.", "")
f2_str = input("f2 = ").replace("np.", "")

f1 = sp.sympify(f1_str)
f2 = sp.sympify(f2_str)

f1_num = sp.lambdify((x_sym, y_sym), f1, "numpy")
f2_num = sp.lambdify((x_sym, y_sym), f2, "numpy")

# Entradas dos dois chutes iniciais
print("Insira o primeiro chute inicial x0 e y0 separados por vírgula:")
x0_str = input()
x0 = np.array([float(v.replace(",", ".")) for v in x0_str.split(",")])

print("Insira o segundo chute inicial x1 e y1 separados por vírgula:")
x1_str = input()
x1 = np.array([float(v.replace(",", ".")) for v in x1_str.split(",")])

tol = float(input("Tolerância desejada: ").replace(",", "."))
max_iter = int(input("Número máximo de iterações: "))

def F(v):
    return np.array([f1_num(v[0], v[1]), f2_num(v[0], v[1])])

for k in range(1, max_iter + 1):
    Fx0 = F(x0)
    Fx1 = F(x1)

    delta_x = x1 - x0
    delta_F = Fx1 - Fx0

    # Aproximação da Jacobiana pelo método das diferenças
    try:
        J_approx = np.outer(delta_F, delta_x) / (delta_x @ delta_x)
    except ZeroDivisionError:
        print("Erro: divisão por zero na aproximação da Jacobiana.")
        break

    # Em sistemas 2x2 a aproximação fica complicada, aqui vamos calcular delta usando pseudo-inversa
    try:
        delta = -np.linalg.pinv(J_approx) @ Fx1
    except np.linalg.LinAlgError:
        print("Erro na inversão da matriz aproximada.")
        break

    x2 = x1 + delta

    erro = np.max(np.abs((x2 - x1) / np.where(x2 != 0, x2, 1)))

    print(f"\nIteração {k}: x = {x2}, erro relativo máximo = {erro:.2e}")

    if erro < tol:
        print(f"\nConvergiu em {k} iterações.")
        print(f"Solução aproximada: x = {x2[0]:.6f}, y = {x2[1]:.6f}")
        break

    x0, x1 = x1, x2
else:
    print("\nNão convergiu no número máximo de iterações.")
```

```python
# Usando o método de Leverrier

import numpy as np

def leverrier_didatico(A):
    n = A.shape[0]
    I = np.eye(n)
    B = A.copy()
    coef = []

    print("=== Método de Leverrier ===")
    print("Matriz A:")
    print(A)
    print()

    for k in range(1, n + 1):
        traco = np.trace(B)
        ak = -traco / k
        coef.append(ak)

        print(f"Passo {k}:")
        print(f"  B_{k} =\n{B}")
        print(f"  Tr(B_{k}) = {traco:.4f}")
        print(f"  a{k} = -Tr(B_{k}) / {k} = {-traco:.4f} / {k} = {ak:.4f}")

        if k < n:
            B = A @ (B + ak * I)
        print()

    # Mostrar o polinômio completo
    print("Polinômio característico:")
    termos = [1] + coef
    grau = n
    polinomio = ""
    for i, a in enumerate(termos):
        exp = grau - i
        if exp > 1:
            polinomio += f"{a:+.4f}λ^{exp} "
        elif exp == 1:
            polinomio += f"{a:+.4f}λ "
        else:
            polinomio += f"{a:+.4f}"
    print("P(λ) =", polinomio)

    return termos

n = int(input("Digite a ordem da matriz quadrada: "))
print("Digite os elementos da matriz linha por linha, separados por vírgula:")

A = []
for i in range(n):
    linha = input(f"Linha {i+1}: ")
    A.append([float(x.replace(",", ".")) for x in linha.split(",")])

A = np.array(A)

# Rodar o método
coeficientes = leverrier_didatico(A)

# Calcular autovalores
autovalores = np.roots(coeficientes)
print("\nAutovalores aproximados (raízes do polinômio):")
for i, val in enumerate(autovalores):
    print(f"λ{i+1} = {val:.6f}")
```

```python
# Usando o Método da Potência

import numpy as np

def metodo_potencia(A, x0, tol=1e-6, max_iter=100):
    x = x0 / np.linalg.norm(x0)
    for i in range(max_iter):
        x_prox = A @ x
        x_prox_norm = np.linalg.norm(x_prox)
        if x_prox_norm == 0:
            print("Erro: vetor nulo encontrado.")
            return None, None
        x_prox = x_prox / x_prox_norm

        # Estimativa do autovalor (Ritz)
        autovalor = np.dot(x.T, A @ x) / np.dot(x.T, x)
        erro = np.linalg.norm(x_prox - x)
        print(f"Iteração {i+1}: λ ≈ {autovalor:.6f}, erro = {erro:.2e}")

        if erro < tol:
            print("\nConvergência alcançada.")
            return autovalor, x_prox

        x = x_prox

    print("\nNão convergiu no número máximo de iterações.")
    return autovalor, x

n = int(input("Digite a ordem da matriz quadrada: "))
print("Digite os elementos da matriz linha por linha, separados por vírgula:")

A = []
for i in range(n):
    linha = input(f"Linha {i+1}: ")
    A.append([float(x.replace(",", ".")) for x in linha.split(",")])
A = np.array(A)

print("\nDigite o vetor inicial (valores separados por vírgula):")
x0 = np.array([float(x.replace(",", ".")) for x in input("x0: ").split(",")])

# Parâmetros de precisão
tol = float(input("Digite a tolerância: ").replace(",", "."))
max_iter = int(input("Digite o número máximo de iterações: "))

# Executar
autovalor, autovetor = metodo_potencia(A, x0, tol, max_iter)

# Mostrar resultados
if autovetor is not None:
    print("\nAutovalor dominante aproximado:", round(autovalor, 6))
    print("Autovetor correspondente (normalizado):")
    for i, val in enumerate(autovetor):
        print(f"x{i+1} = {val:.6f}")
```

```python
# Usando o Método da Potência Inversa

import numpy as np

def metodo_potencia_inversa(A, x0, tol=1e-6, max_iter=100):
    x = x0 / np.linalg.norm(x0)
    for i in range(max_iter):
        try:
            z = np.linalg.solve(A, x)  # em vez de A⁻¹ @ x
        except np.linalg.LinAlgError:
            print("Erro: matriz A é singular (não invertível).")
            return None, None

        z_norm = np.linalg.norm(z)
        if z_norm == 0:
            print("Erro: vetor nulo encontrado.")
            return None, None

        x_prox = z / z_norm
        autovalor = np.dot(x_prox.T, A @ x_prox) / np.dot(x_prox.T, x_prox)
        erro = np.linalg.norm(x_prox - x)

        print(f"Iteração {i+1}: λ ≈ {autovalor:.6f}, erro = {erro:.2e}")

        if erro < tol:
            print("\nConvergência alcançada.")
            return autovalor, x_prox

        x = x_prox

    print("\nNão convergiu no número máximo de iterações.")
    return autovalor, x

n = int(input("Digite a ordem da matriz quadrada: "))
print("Digite os elementos da matriz linha por linha, separados por vírgula:")

A = []
for i in range(n):
    linha = input(f"Linha {i+1}: ")
    A.append([float(x.replace(",", ".")) for x in linha.split(",")])
A = np.array(A)

print("\nDigite o vetor inicial (valores separados por vírgula):")
x0 = np.array([float(x.replace(",", ".")) for x in input("x0: ").split(",")])

# Parâmetros de precisão
tol = float(input("Digite a tolerância: ").replace(",", "."))
max_iter = int(input("Digite o número máximo de iterações: "))

# Executar o método
autovalor, autovetor = metodo_potencia_inversa(A, x0, tol, max_iter)

# Resultados
if autovetor is not None:
    print("\nAutovalor de menor módulo aproximado:", round(autovalor, 6))
    print("Autovetor correspondente (normalizado):")
    for i, val in enumerate(autovetor):
        print(f"x{i+1} = {val:.6f}")
```

```python
# Aplicando um Ajuste Linear Pelo Método de Minímos Quadrados

import numpy as np

def ajuste_linear_minimos_quadrados(x, y):
    if len(x) != len(y):
        raise ValueError("Os vetores x e y devem ter o mesmo tamanho.")

    n = len(x)
    somax = np.sum(x)
    somay = np.sum(y)
    somax2 = np.sum(x**2)
    somaxy = np.sum(x * y)

    a = (n * somaxy - somax * somay) / (n * somax2 - somax**2)
    b = (somay - a * somax) / n

    return a, b

# Entrada do usuário
print("Insira os valores de x separados por vírgula:")
x = np.array([float(valor.strip().replace(",", ".")) for valor in input().split(",")])

print("Insira os valores de y separados por vírgula:")
y = np.array([float(valor.strip().replace(",", ".")) for valor in input().split(",")])

# Cálculo
a, b = ajuste_linear_minimos_quadrados(x, y)

# Exibição dos coeficientes
print(f"\nCoeficiente angular (a): {a:.6f}")
print(f"Coeficiente linear (b): {b:.6f}")

# Exibição da equação ajustada como função
sinal = "+" if b >= 0 else "-"
print(f"\nFunção ajustada:")
print(f"f(x) = {a:.4f}x {sinal} {abs(b):.4f}")
```

```python
# Aplicando Ajuste Com a Linearização de Equações Não Lineares

import numpy as np

def ajuste_exponencial(x, y):
    if np.any(y <= 0):
        raise ValueError("Todos os valores de y devem ser positivos para ajuste exponencial.")

    Y = np.log(y)  # Lineariza o y
    n = len(x)

    somax = np.sum(x)
    somay = np.sum(Y)
    somax2 = np.sum(x**2)
    somaxy = np.sum(x * Y)

    b = (n * somaxy - somax * somay) / (n * somax2 - somax**2)
    A = (somay - b * somax) / n
    a = np.exp(A)  # Volta para a escala original

    return a, b

# Entrada do usuário
print("Insira os valores de x separados por vírgula:")
x = np.array([float(valor.strip().replace(",", ".")) for valor in input().split(",")])

print("Insira os valores de y separados por vírgula (todos positivos):")
y = np.array([float(valor.strip().replace(",", ".")) for valor in input().split(",")])

# Cálculo
try:
    a, b = ajuste_exponencial(x, y)
    print(f"\nFunção ajustada:")
    print(f"f(x) = {a:.4f} * exp({b:.4f} * x)")
except ValueError as e:
    print("Erro:", e)
```

```python
# Aplicando Interpolação Pelo Metódo de Vandermonde

import numpy as np

def interpolacao_vandermonde(x_pontos, y_pontos):
    V = np.vander(x_pontos, increasing=True)
    a = np.linalg.solve(V, y_pontos)
    return a

def avaliar_polinomio(a, x):
    return sum(a[i] * x**i for i in range(len(a)))

def montar_polinomio_simbolico(coef):
    termos = []
    for i, a in enumerate(coef):
        termo = f"{a:.6f}"
        if i == 1:
            termo += "x"
        elif i >= 2:
            termo += f"x^{i}"
        termos.append(termo)
    return " + ".join(termos).replace("+ -", "- ")

# Entrada dos dados
print("Digite os valores de x (separados por vírgula):")
x_pontos = np.array([float(v.strip().replace(",", ".")) for v in input().split(",")])

print("Digite os valores de y (separados por vírgula):")
y_pontos = np.array([float(v.strip().replace(",", ".")) for v in input().split(",")])

# Solicita o ponto a ser interpolado
x_avaliar = float(input("\nDigite o valor de x para calcular p(x): ").replace(",", "."))

# Processamento
coeficientes = interpolacao_vandermonde(x_pontos, y_pontos)
p_x = avaliar_polinomio(coeficientes, x_avaliar)
polinomio = montar_polinomio_simbolico(coeficientes)

# Saída
print("\nPolinômio interpolador encontrado:")
print(f"p(x) = {polinomio}")
print(f"\nValor de p({x_avaliar}) = {p_x:.6f}")
```

```python
# Aplicando Interpolação Pelo Metódo de Lagrange

import sympy as sp

def lagrange_simbolico(x_vals, y_vals):
    x = sp.Symbol('x')
    n = len(x_vals)
    p_x = 0

    for i in range(n):
        termo = y_vals[i]
        for j in range(n):
            if i != j:
                termo *= (x - x_vals[j]) / (x_vals[i] - x_vals[j])
        p_x += termo

    return sp.simplify(p_x)

# Entrada dos dados
print("Digite os valores de x (separados por vírgula):")
x_pontos = [float(v.strip().replace(",", ".")) for v in input().split(",")]

print("Digite os valores de y (separados por vírgula):")
y_pontos = [float(v.strip().replace(",", ".")) for v in input().split(",")]

# Construção do polinômio
polinomio = lagrange_simbolico(x_pontos, y_pontos)

# Exibição do polinômio
print("\nPolinômio interpolador p(x):")
print(f"p(x) = {polinomio}")

# Pergunta opcional
resp = input("\nDeseja calcular p(x) para algum valor? (s/n): ").lower()
if resp == 's':
    valor = float(input("Digite o valor de x: ").replace(",", "."))
    resultado = polinomio.evalf(subs={"x": valor})
    print(f"p({valor}) = {resultado:.6f}")
```

```python
# Aplicando Interpolação Pelo Metódo de Newton

import sympy as sp

def interpolacao_newton(x_vals, y_vals):
    x = sp.Symbol('x')
    n = len(x_vals)
    f = 0
    coef = [0] * n

    # Cálculo das diferenças divididas
    for i in range(n):
        coef[i] = y_vals[i]
    for j in range(1, n):
        for i in range(n-1, j-1, -1):
            coef[i] = (coef[i] - coef[i-1]) / (x_vals[i] - x_vals[i-j])

    # Construção do polinômio
    termo = 1
    for i in range(n):
        f += coef[i] * termo
        termo *= (x - x_vals[i])

    return sp.simplify(f)

# Entrada do usuário
print("Digite os valores de x (separados por vírgula):")
x_vals = [float(num.replace(",", ".")) for num in input().split(",")]

print("Digite os valores de y (separados por vírgula):")
y_vals = [float(num.replace(",", ".")) for num in input().split(",")]

# Gerar polinômio
polinomio = interpolacao_newton(x_vals, y_vals)

# Mostrar resultado
print("\nPolinômio interpolador p(x):")
sp.pprint(polinomio, use_unicode=True)

# Avaliação opcional
resposta = input("\nDeseja calcular p(x) para algum valor? (s/n): ").lower()
if resposta == 's':
    x_input = float(input("Digite o valor de x: ").replace(",", "."))
    resultado = polinomio.subs("x", x_input)
    print(f"\np({x_input}) = {resultado.evalf()}")
```

```python
# Trabalhando com Derivada Por Diferenças Fnitas a Partir de 3 Pontos

import sympy as sp

def encontrar_indice(x_list, valor):
    """Encontra o índice do ponto mais próximo de valor"""
    try:
        return x_list.index(valor)
    except ValueError:
        diffs = [abs(x - valor) for x in x_list]
        return diffs.index(min(diffs))

def derivada_avancada(x, y, i):
    h = x[i+1] - x[i]
    return (y[i+1] - y[i]) / h

def derivada_atrasada(x, y, i):
    h = x[i] - x[i-1]
    return (y[i] - y[i-1]) / h

def derivada_central(x, y, i):
    h = x[i+1] - x[i-1]
    return (y[i+1] - y[i-1]) / h

# Entrada da função
x_sim = sp.Symbol('x')
f_expr_str = input("Digite a função f(x): ")
f_expr = sp.sympify(f_expr_str)

# Entrada dos dados da tabela
print("Digite os valores de x separados por vírgula:")
x_vals = [float(v.strip().replace(",", ".")) for v in input().split(",")]

print("Digite os valores de f(x) separados por vírgula:")
y_vals = [float(v.strip().replace(",", ".")) for v in input().split(",")]

if len(x_vals) != len(y_vals):
    print("Os vetores x e y devem ter o mesmo tamanho.")
    exit()

# Derivada simbólica
f_derivada = sp.diff(f_expr, x_sim)

# Ponto a ser avaliado
xp = float(input("Digite o ponto x onde deseja calcular a derivada: ").replace(",", "."))
valor_exato = float(f_derivada.evalf(subs={x_sim: xp}))

# Identificar índice mais próximo
i = encontrar_indice(x_vals, xp)

# Cálculo das aproximações e ERe
print(f"\nDerivada exata de f(x) = {f_expr} no ponto x = {xp}: f'(x) = {f_derivada} ≈ {valor_exato:.6f}")
print("\nAproximações numéricas:")

resultados = {}

if i == 0:
    aprox = derivada_avancada(x_vals, y_vals, i)
    resultados["avançado"] = aprox
    print(f"Avançado: {aprox:.6f}")
elif i == len(x_vals) - 1:
    aprox = derivada_atrasada(x_vals, y_vals, i)
    resultados["atrasado"] = aprox
    print(f"Atrasado: {aprox:.6f}")
else:
    avan = derivada_avancada(x_vals, y_vals, i)
    atra = derivada_atrasada(x_vals, y_vals, i)
    cent = derivada_central(x_vals, y_vals, i)
    resultados["avançado"] = avan
    resultados["atrasado"] = atra
    resultados["central"] = cent
    print(f"Avançado: {avan:.6f}")
    print(f"Atrasado: {atra:.6f}")
    print(f"Central: {cent:.6f}")

# Erro Relativo Exato
print("\nErro Relativo Exato (ERe):")
for metodo, aprox in resultados.items():
    ere = abs((valor_exato - aprox) / valor_exato)
    print(f"{metodo.capitalize()}: {ere:.6f}")
```

```python
# Aplicando a Regra dos Trapézios para Integração Númerica

import sympy as sp
import numpy as np

# Entrada da função
x = sp.Symbol('x')
f_str = input("Digite a função f(x): ").lower().replace("^", "**").replace("e**", "exp")
f = sp.sympify(f_str, locals={"exp": sp.exp, "ln": sp.log})
f_lambd = sp.lambdify(x, f, "numpy")

# Limites de integração e número de subintervalos
a = float(input("Digite o limite inferior a: ").replace(",", "."))
b = float(input("Digite o limite superior b: ").replace(",", "."))
n = int(input("Digite o número de subintervalos (n): "))

# Cálculo da Regra do Trapézio
h = (b - a) / n
x_vals = np.linspace(a, b, n+1)
y_vals = f_lambd(x_vals)

area = (h / 2) * (y_vals[0] + 2 * sum(y_vals[1:-1]) + y_vals[-1])

# Resultado
print(f"\nIntegral aproximada de f(x) no intervalo [{a}, {b}] usando a Regra dos Trapézios com n={n}:")
print(f"Resultado ≈ {area:.6f}")
```

```python
# Aplicando a Regra de Simpson para Integração Númerica

import sympy as sp
import numpy as np

# Entrada da função
x = sp.Symbol('x')
f_str = input("Digite a função f(x): ").lower().replace("^", "**").replace("e**", "exp")
f = sp.sympify(f_str, locals={"exp": sp.exp, "ln": sp.log})
f_lambd = sp.lambdify(x, f, "numpy")

# Limites de integração e número de subintervalos
a = float(input("Digite o limite inferior a: ").replace(",", "."))
b = float(input("Digite o limite superior b: ").replace(",", "."))
n = int(input("Digite o número de subintervalos (n, precisa ser par): "))

# Validação
if n % 2 != 0:
    print("\nO número de subintervalos (n) precisa ser PAR para a Regra de Simpson 1/3.")
else:
    h = (b - a) / n
    x_vals = np.linspace(a, b, n + 1)
    y_vals = f_lambd(x_vals)

    soma_pares = sum(y_vals[2:n:2])  # índices pares (exceto o primeiro e último)
    soma_impares = sum(y_vals[1:n:2])  # índices ímpares

    integral = (h / 3) * (y_vals[0] + 4 * soma_impares + 2 * soma_pares + y_vals[-1])

    print(f"\nIntegral aproximada de f(x) no intervalo [{a}, {b}] usando Simpson 1/3 com n={n}:")
    print(f"Resultado ≈ {integral:.6f}")
```

```python
# Aplicando a Regra de Simpson Repetido para Integração Númerica

import sympy as sp
import numpy as np

# Entrada da função
x = sp.Symbol('x')
f_str = input("Digite a função f(x): ").lower().replace("^", "**").replace("e**", "exp")
f = sp.sympify(f_str, locals={"exp": sp.exp, "ln": sp.log})
f_lambd = sp.lambdify(x, f, "numpy")

# Limites de integração e número de subintervalos
a = float(input("Digite o limite inferior a: ").replace(",", "."))
b = float(input("Digite o limite superior b: ").replace(",", "."))
n = int(input("Digite o número de subintervalos (n - deve ser par): "))

# Verifica se n é par
if n % 2 != 0:
    raise ValueError("O número de subintervalos (n) deve ser par para usar a Regra de Simpson.")

# Cálculo
h = (b - a) / n
x_vals = np.linspace(a, b, n + 1)
y_vals = f_lambd(x_vals)

# Aplicando a Regra de Simpson Composta
soma = y_vals[0] + y_vals[-1]
soma += 4 * sum(y_vals[1:-1:2])  # Índices ímpares
soma += 2 * sum(y_vals[2:-2:2])  # Índices pares
integral = (h / 3) * soma

# Resultado
print(f"\nIntegral aproximada de f(x) no intervalo [{a}, {b}] usando a Regra de Simpson Repetida com n={n}:")
print(f"Resultado ≈ {integral:.6f}")
```
